```python
from EnergyDisaggregation.energydisaggregation.models.nn_model import NeuralNetworkModel, preprocess
```

```python
import pandas as pd
```

```python
df = pd.read_csv('./EnergyDisaggregation/energydisaggregation/data_storage/df_process.csv')
```

```python
df = df[df['Région']=="Provence-Alpes-Côte d'Azur"]
```

```python
df.head()
```

```python
df_preprocess = preprocess(df)
```

```python
df_preprocess.head()
```

```python
df_train = df_preprocess[df_preprocess.index.year <= 2017]
df_test = df_preprocess[df_preprocess.index.year == 2018]
```

```python
nn_model = NeuralNetworkModel(hidden_dim=25, output_dim=1, num_layers=5)
```

```python
print(nn_model.model)
```

```python
nn_model.fit(df_train)
```

```python
a, b, c = nn_model.predict(df_test)
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'c':a,'t':b,'r':c})
```

```python
df.to_csv('df_test_NN.csv')
```

```python
plt.plot(a, label='Consommation prédite')
plt.plot(b, label='Conso Thermosensible')
plt.plot(c, label='Consommation régulière')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
nn_model.score(df_test['consommation brute électricité (mw) - rte'], a)
```
