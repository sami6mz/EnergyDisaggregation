```python
data_path = 'data_storage/raw_data/consommation-quotidienne-brute-regionale.csv'
df2 = pd.read_csv(data_path, delimiter=';')
```

```python
#TRANSFORMER LES VALEURS NaN de la consommation gaz en 0
df2["Consommation brute gaz (MW PCS 0°C) - GRTgaz"].fillna(0,inplace=True)
```

```python
df2_sorted = df2.sort_values(by=['Date - Heure', 'Région'])
```

```python
#print(df2_sorted)
```

```python
conso_totale = df2_sorted['Consommation brute totale (MW)']
df.rename(columns={'consommation brute électricité (mw) - rte': 'consommation brute totale (mw)'}, inplace=True)
```

```python
df['consommation brute totale (mw)'] = conso_totale
```

```python
df['ratio gaz electricite'] = df2_sorted['Consommation brute gaz totale (MW PCS 0°C)'] / df2_sorted['Consommation brute électricité (MW) - RTE'] 
```

```python
print(df.columns)
```

```python
conso_totale = df['consommation brute totale (mw)']
lag = 7
for i in range(lag):
    conso_lag = conso_totale - conso_totale.shift(-12*(i+1))
    df['consommation brute électricité (mw) - rte_lag_{i}'] = conso_lag
```

```python
#print(df['consommation brute électricité (mw) - rte_lag_1'].head(40))
```

```python
df.rename(columns={
    'consommation brute électricité (mw) - rte_lag_1': 'consommation brute totale (mw) - lag_1',
    'consommation brute électricité (mw) - rte_lag_2': 'consommation brute totale (mw) - lag_2',
    'consommation brute électricité (mw) - rte_lag_3': 'consommation brute totale (mw) - lag_3',
    'consommation brute électricité (mw) - rte_lag_4': 'consommation brute totale (mw) - lag_4',
    'consommation brute électricité (mw) - rte_lag_5': 'consommation brute totale (mw) - lag_5',
    'consommation brute électricité (mw) - rte_lag_6': 'consommation brute totale (mw) - lag_6',
    'consommation brute électricité (mw) - rte_lag_7': 'consommation brute totale (mw) - lag_7'
}, inplace=True)
```

```python
df.drop(columns=['consommation brute électricité (mw) - rte_lag_{i}'], inplace=True)
print(df.columns)
```

```python
df.to_csv('df_processed_ratio.csv', index=False)
```
